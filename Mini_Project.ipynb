```python
df1['Value']
```

```python
#Removing the Euro symbol as this is an Numerical column and replacing the Million(M) or Thousand(K) with required values.
```

```python
df1['Value']=df1['Value'].str.replace('€','').str.replace('M','*1000000').str.replace('K','*1000')
```

```python
df1['Value']
```

```python
#In python we have eval function - learnt about this from towards data scienece website.
#It will take the argument value and reads it. When it finds a numerical symbol between two values, eval will perform
#that opeartion in the arguments value. 
```

```python
def eval_func(x):
    if type(x)==str:
        return eval(x)
```

```python
df1['Value']=df1['Value'].apply(eval_func)
```

```python
df1['Value']
```

```python
#so we use eval function apply to the Wage column also
```

```python
df1['Wage'] 
```

```python
df1['Wage']=df1['Wage'].str.replace('€','').str.replace('K','*1000')
df1['Wage']
```

```python
df1['Wage']=df1['Wage'].apply(eval_func)
```

```python
df1['Wage']
```

```python
df1['Joined']
```

```python
#INFERENCE

#This one has null values and when trying to convert will show in float  or when doing other way using pd.todatetime
#will end up showing 1970 as default for all

#We will need to deal with the NaN values before converting it. So We will be delaing it below and convert it
```

```python
df1['Height']=df1['Height'].str.replace('\'','.').astype(float)
```

```python
df1['Weight']=df1['Weight'].str.replace('lbs','').astype(float)
```

```python
df1['Release Clause']=df1['Release Clause'].str.replace('€','').str.replace('M','*1000000').str.replace('K','*1000')
```
